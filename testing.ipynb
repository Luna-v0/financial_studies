```python
import tabula
import pandas as pd
import re

# Read pdf into list of DataFrame
dfs = tabula.read_pdf("ignore/extrato.pdf", pages='all')

df = dfs[2]
first_row = pd.DataFrame([df.columns], columns=df.columns)

# Step 2: Append the original DataFrame below
df = pd.concat([first_row, df], ignore_index=True)

# Step 3: Reset proper column names
df.columns = [f"Column{i+1}" for i in range(df.shape[1])]
```

```python

def join_alternating_rows(input_df):
    current_df = input_df.copy()
    
    rows = []
    
    nextRow = False
    for _, row in current_df.iterrows():
        if pd.isna(row[0]):
            continue
        
        if nextRow:
            nextRow = False
            rows[len(rows)-1] = rows[len(rows)-1] + list(row)
            continue
        
        if type(row[0]) == str and re.search('\d\d/\d\d',row[0]):
            rows.append(list(row))
            nextRow = True
            
    return pd.DataFrame(rows)

df
```

```python
def half_df(input_df):
    current_df = input_df.copy()
    
    mid = len(current_df.columns) // 2
    

    # Split into two DataFrames (first half and second half)
    df1 = current_df.iloc[:, :mid].reset_index(drop=True)
    df2 = current_df.iloc[:, mid:].reset_index(drop=True)

    df1.columns = [f"Column_{i+1}" for i in range(len(df1.columns))]
    df2.columns = [f"Column_{i+1}" for i in range(len(df2.columns))]

    # Append one below the other
    appended_df = pd.concat([df1, df2], axis=0, ignore_index=True)
    
    return appended_df
```

```python
def drop_column_by_number(input_df,column_num:list[int] | int):
    current_df = input_df.copy()
    if type(column_num) == int:
        current_df.drop(current_df.columns[column_num], axis=1, inplace=True)
    else:
        sorted_columns = sorted(column_num)
        for x in range(len(column_num)):
            index = sorted_columns[x] - x
            current_df = drop_column_by_number(current_df,index)
    return current_df

drop_column_by_number(join_alternating_rows(df),[1,3,4,6,7,8,9])
```

```python
df1 = drop_column_by_number(join_alternating_rows(half_df(dfs[1])),3)
df2 = drop_column_by_number(join_alternating_rows(df),[1,3,4,6,7,8,9])

df1.columns = [f"Column_{i+1}" for i in range(len(df1.columns))]
df2.columns = [f"Column_{i+1}" for i in range(len(df2.columns))]

final_df = pd.concat([df1,df2])
final_df.columns = [f"Column_{i+1}" for i in range(len(df2.columns))]
```
